```python
# Method 1 - Create a Series
array = np.arange(5)
pd.Series(array)
```

```python
# Method 2 - Create a Series
series = pd.Series(np.arange(9), name="Test_Num")
```

```python
# Remember that Series are 1-Dimensional data
pd.Series(np.arange(10).reshape(2, 5), name="New_Series")
```

```python
series.values
```

```python
series.values.mean()
```

```python
series.mean()
```

```python
np.mean(series)
```
