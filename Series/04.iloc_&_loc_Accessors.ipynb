# THE `.iloc[]` AND `.loc[]` ACCESSORS

---

## The `.iloc[]` Method

> The `.iloc[]` method is the preferred way to access values by their positional index
>
>   - This method works even when Series have a custom, non integer index
>   - It is more efficient than slicing and is recommended py Pandas creators

```python
import numpy as np
import pandas as pd
```

```python
sales_series = pd.Series(
    [0, 5, 155, 0, 518],
    index=["coffee", "bananas", "tea", "coconut", "sugar"],
    dtype="int",
)
sales_series
```

```python
sales_series.iloc[2]
```

```python
sales_series.iloc[2:4]
```

```python
my_series = pd.Series(range(5), index=["day 0", "day 1", "day 2", "day 3", "day 4"])
my_series
```

```python
my_series.iloc[0]
```

```python
my_series.iloc[-1]
```

```python
my_series.iloc[[1, 3, 4]]
```

```python
my_series.iloc[:3]
```

---

## The `.loc[]` Method


>The `.loc[]` method is the preferred way to access values by their custom labels
>
>---
>>The `.loc[]` method works even when the indices are integers, but if they are custom integers nor ordered from 0 to n-1, the rows will be returned based on the **labels** themselves and NOT by their numeric position

```python
sales_series
```

```python
sales_series.loc["tea"]
```

```python
sales_series.loc["bananas":"coconut"]
```

```python
my_series
```

```python
my_series.loc["day 2"]
```

```python
my_series.loc["day 1":"day 3"]
```

```python
my_series.index = [0, 2, 3, 100, 5]
my_series
```

```python
my_series.loc[:5]
```

```python
my_series.reset_index(drop=True)
```

```python
my_series.loc[:3]
```

```python
my_series.loc[:5]
```
