# FILTERING SERIES & LOGICAL TESTS

---

>
>|**DESCRIPTION**|**Python Operator**|**Pandas Method**|
>|:---|:---|:---|
>|**Equal**|==|`.eq()`|
>|**Not Equal**|!=|`.ne()`|
>|**Less Than or Equal**|<=|`.le()`|
>|**Less Than**|<|`.lt()`|
>|**Greater Than or Equal**|>=|`.ge()`|
>|**Greater Than**|>|`.gt()`|
>|**Membership Test**|in|`.isin()`|
>|**Inverse Membership Test**|not in|`~.isin()`|

```python
import numpy as np
import pandas as pd
```

```python
sales_series = pd.Series(
    data=[0, 5, 155, 0, 518],
    index=["coffee", "coffee", "tea", "coconut", "sugar"],
    name="Sales",
)
sales_series
```

```python
# it returns all ros with a value greater than 0
sales_series[sales_series > 0]
```

```python
# returns all rows with a value greater than 0 AND index value equal to 'coffee'
mask = (sales_series > 0) & (sales_series.index == "coffee")
sales_series[mask]
```

---

```python
sales_series == 5
```

```python
sales_series.eq(5)
```

```python
# Membership Tests
sales_series.index.isin(["coffee", "tea"])
```

```python
# Inverse Membership Tests
~sales_series.index.isin(["coffee", "tea"])
```

```python
sales_series[sales_series == 5]
```

```python
sales_series[~(sales_series == 5)]
```

```python
# Membership Test
sales_series[sales_series.isin([0, 518])]
```

```python
# Invert Membership Test
sales_series[~sales_series.isin([0, 518])]
```

```python
# Greater Than - Python Operator
sales_series[sales_series > 5]
```

```python
# Greater Than - Pandas Method
sales_series[sales_series.gt(5)]
```

```python
# Not Greater Than - Pandas Method
sales_series[~(sales_series.gt(5))]
```

```python
sales_series[sales_series.isin([0, 5])]
```
