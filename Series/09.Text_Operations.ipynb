# TEXT SERIES OPERATIONS

---

>The Pandas `str` accessor lets you access many string methods
>  - These methods all return a Series (split returns multiple series)

<br>

>|**STRING METHOD**|**DESCRIPTION**|
>|:---|:---|
>|**.strip(), .lstrip(), .rstrip()**|Removes all leading and/or trailing characters (spaces by default)|
>|**.upper(), .lower()**|Converts all characters to upper or lower case|
>|**.slice(start:stop:step)**|Applies a slice to the strings in a Series|
>|**.count("string")**|Counts all instances of a given string|
>|**.contains("string")**|Returns `True` if a given string is found; `False` if not|
>|**.replace("a", "b")**|Replaces instances of string "a" with string "b"|
>|**.split("delimiter", expand=True)**|Splits strings based on a given delimiter string, and returns a DataFrame with a Series for each split|
>|**.len()**|Returns the length of each string in a Series|
>|**.startswith("string", .endswith("string")**)|Returns `True` if a string starts or ends with given string; `False` if not|

<br>

```python
import numpy as np
import pandas as pd
```

```python
prices = pd.Series(data=["$3.99", "$5.99", "$22.99", "$7.99", "$33.99"])
prices
```

```python
# Look for the string '3'
prices.str.contains("3")
```

```python
# Remove '$' and convert numbers into floats
clean = prices.str.strip("$").astype("float")
print(f"Shape of the data: {clean.shape}\n")
clean
```

```python
string_series = pd.Series(data=["day 0", "day 1", "day 2", "day 3", "day 4"])
string_series
```

```python
# using 'contains' methods
string_series.str.contains("1")
```

```python
# Using 2 Methods on a single line of code
string_series.str.upper().str.contains("A")
```

```python
# Cleaning 'day ' from each row and converting into int (METHOD 1)
string_series.str.strip("day ").astype("int")
```

```python
# Cleaning 'day ' from each row and converting into int  (METHOD 2)
string_series.str.slice(-1).astype("int")
```

```python
# Cleaning 'day ' from each row and converting into int  (METHOD 3)
string_series.str[-1].astype("int")
```

```python
# Slicing - Method 1
string_series.str.slice(1, 3)
```

```python
# Slicing - Method 2
string_series.str[1:3]
```

```python
# Splitting with 'expand=False' - It is a Series
string_series.str.split(" ", expand=False)
```

```python
# Splitting with 'expand=True' - It is a DataFrame
string_series.str.split(" ", expand=True)
```

```python
# Provided series with combined age and label data
age_data = pd.Series(["Adult 25", "Child 12", "Adult 64", "Teen 17", "Adult 45"])

# Split the string to separate the age from the labels
split_data = age_data.str.split(" ", expand=True)
split_data
```

```python
split_data[1] = split_data[1].astype("int")
split_data
```

# ASSIGNMENT: SERIES OPERATIONS

---

* Increase the prices in the oil series by 10%, and add an additional 2 dollars per barrel on top of that.

* Then, create a series that represents the difference between each price and max price.

* Finally, extract the month from the string dates in the index and store them as an integer in their own series.

```python
oil = pd.read_csv("Data/oil.csv", index_col="date").dropna()

print(f"  - Data Shape: {oil.shape}\n  - Data Type: {type(oil)}\n")
oil.head()
```

```python
oil = oil["dcoilwtico"].rename(index="price")

print(f"  - Data Shape: {oil.shape}\n  - Data Type: {type(oil)}\n")
oil.head()
```

```python
oil = oil[1000:1100]
oil
```

```python
# Increase the prices in the oil series by 10%, and add an additional 2 dollars per barrel on top of that.
oil_percent = round(oil.mul(1.1), 3).add(2)
oil_percent
```

```python
# Then, create a series that represents the difference between each price and max price.
max_price = oil.max()
print(max_price, "\n")
oil_diff = oil.sub(max_price).div(max_price).round(4)
oil_diff
```

```python
# Finally, extract the month from the string dates in the index and store them as an integer in their own series.
month = oil.index.str[5:7].astype("int")
print(month, "\n\n")

month_series = pd.Series(data=month)
month_series
```
