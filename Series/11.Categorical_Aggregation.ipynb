# CATEGORICAL SERIES AGGREGATION

---

>You can use this methods to **aggregate and analyze Categorical Series**

<br>

>|**METHOD**|**DESCRIPTION**|
>|:---|:---|
>|**.unique()**|Returns an array of unique items in a Series|
>|**.nunique()**|Returns the number of unique items in a Series|
>|**.value_counts()**|Returns a Series of unique items and their frequency|


<br>

```python
import numpy as np
import pandas as pd
```

```python
items = pd.Series(data=['coffee', 'coffee', 'tea', 'coconut', 'sugar'])
items
```

```python
items.value_counts()
```

```python
items.value_counts(normalize=True)
```

```python
items.unique()
```

```python
items.nunique()
```

# ASSIGNMENT 5: SERIES AGGREGATIONS

---

```python
oil = pd.read_csv('Data/oil.csv', index_col='date').dropna()

print(f'Oil data type: {type(oil)}   -   {oil.shape}\n')
oil.head()
```

```python
oil_series = oil['dcoilwtico'].rename(index='price')[1000:1100]
oil_series
```

```python
# Calculate the sum and mean of prices in the month of March. 
oil_march = oil_series[oil_series.index.str[5:7] == '03']
oil_march_mean = oil_march.mean()
oil_march_sum = oil_march.sum()

print(f'March Mean: {oil_march_mean:.4f}   - March Sum: {oil_march_sum:.4f}\n')
```

```python
# Next, calculate how many prices were recorded in January and February.
jan_oil = oil_series[oil_series.index.str[5:7] == '01']
jan_count = jan_oil.count()

feb_oil = oil_series[oil_series.index.str[5:7] == '02']
feb_count = feb_oil.count()

print(f"Total prices on January: {jan_count}\nTotal prices on February: {feb_count}\n")
```

```python
# Then, calculate the 10th and 90th percentiles across all data.
oil_percentiles = oil_series.quantile([0.1, 0.9])
print(oil_percentiles, '\n')
```

```python
# Finally, how often did integer dollar value (e.g. 51, 52) occur in the data? Normalize this to a percentage.   
oil_series.astype('int').value_counts(normalize=True)
```
