# 1. Apply Method




The `.apply()` method lets you apply custom functions to Pandas Series

- This function `will not be vectorized`, so it's not as efficient as native functions

```python
import numpy as np
import pandas as pd
import time
```

```python
clean_wholesale = pd.Series([3.99, 5.99, 22.99, 7.99, 33.99])
clean_wholesale
```


---

```python
def discount(price):
    if price > 20:
        return round(price * 0.9, 2)
    return price
```

```python
%%timeit -r 5 -n 1000
clean_wholesale.apply(discount)
```

```python
%%timeit -r 5 -n 1000
clean_wholesale.apply(lambda x: round(x * 0.9, 2) if x > 20 else x)
```

# 2. LABORATORY:

```python
str_series = pd.Series(['day 0', 'day 0', 'day 2', 'day 2', 'day 4'])
str_series
```

```python
%%timeit -r 5 -n 1000
str_series.apply(lambda x: x[-1])
```

```python
%%timeit -r 5 -n 1000
str_series.str[-1]
```

---

```python
def search(string, looking_for):
    if looking_for in string:
        return 'Found It!'
    return 'Nope'
```

```python
str_series.apply(search, args='2')
```

```python
str_series.apply(search, args='d')
```
