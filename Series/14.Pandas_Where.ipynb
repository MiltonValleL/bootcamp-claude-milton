# PANDAS WHERE vs. NUMPY WHERE

---

Pandas `.where()` method returns series values based on a logical condition

```python
df.where(logical_test, value if False, inplace=False)
```

````python
import numpy as np
import pandas as pd
````

````python
clean_wholesale = pd.Series([3.99, 5.99, 22.99, 7.99, 33.99])
clean_wholesale
````

````python
# Working with Pandas '.where()'
clean_wholesale.where(clean_wholesale <=20, round(clean_wholesale * 0.9, 2))
````

````python
# Working with Pandas '.where()'
clean_wholesale.where(~(clean_wholesale > 20), round(clean_wholesale * 0.9, 2))
````

<br><br>

---

````python
# Working with NUMPY 'np.where()'
np.where(clean_wholesale > 20, 'Discounted', 'Regular Price')
````

# ASSIGNMENT: APPLY & WHERE

````python
oil = pd.read_csv('Data/oil.csv', index_col='date').dropna()
oil = oil['dcoilwtico'].rename('Price')[1000:1100]
oil
````

````python
oil_series = oil.where(~oil.isin([51.44, 47.83]), np.nan)
oil_series
````

````python
oil_series.isna().sum()
````

````python
#Write a function that outputs ‘buy’ if price is less than the 90th percentile and ‘wait’ if it’s not. Apply it to the oil series.
def buy_bool(price, limit):
    if price < limit:
        return 'Buy'
    return 'Wait'
````

````python
oil_series.apply(buy_bool, args=(oil_series.quantile(0.9),))
````

---

````python
oil_series.apply(lambda x: 'Buy' if x < oil_series.quantile(0.9) else 'Wait')
````

````python
#Then, create a series that multiplies price by .9 if the date is ‘2016-12-23’ or ‘2017-05-10’, and 1.1 for all other dates. 
np.where(oil_series.index.isin(['2016-12-23', '2017-05-10']),oil_series * 0.9, oil_series * 1.1)
````

````python
df = pd.DataFrame(oil_series)
df['new_price'] = np.where(oil_series.index.isin(['2016-12-23', '2017-05-10']), oil_series * 0.9, oil_series * 1.1)
df
````
