# ASSIGNMENT 1: SERIES BASICS

---

```python
oil = pd.read_csv("Data/oil.csv").dropna()

oil_array = np.array(oil["dcoilwtico"].iloc[1000:1100])
oil_array
```

```python
oil_series = pd.Series(oil_array, name="oil_price")

display(oil_series)
print(f"\nName: {oil_series.name}")
print(f"Data Type: {oil_series.dtype}")
print(f"Size: {oil_series.size}")
print(f"Index: {oil_series.index}")
print(f"Series Mean: {oil_series.mean():.4f}")

int_oil_series = oil_series.astype("int")
print(f"Integer Mean: {int_oil_series.mean():.4f}")
```
