# SORTING SERIES

---

> We can sort Series by their values or their index

```python
import numpy as np
import pandas as pd
```

```python
sales_series = pd.Series(
    data=[0, 5, 155, 0, 518],
    index=["coffee", "coffee", "tea", "coconut", "sugar"],
    name="Sales",
)
sales_series
```

---

## Sorting Values

```python
sales_series.sort_values()
```

```python
sales_series.sort_values(ascending=False)
```

---

## Sorting Index

```python
sales_series.sort_index()
```

```python
sales_series.sort_index(ascending=False)
```
