# HEATMAPS AND PIVOT TABLES

---

You can style a DataFrame based on its values to create a **heatmap**

 - Simply chain `.style.background_gradient()` to you DataFrame and add a `cmap` argument

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')

reduced_retail = (retail[(retail.store_nbr <= 4) & 
    (retail.family.isin(['AUTOMOTIVE', 
                         'BOOKS', 
                         'HARDWARE', 
                         'MAGAZINES']))])
reduced_retail
```

---

## 1. Heatmap for all the table

```python
(reduced_retail.pivot_table(index='family', 
                           columns='store_nbr', 
                           values='sales',
                           aggfunc=('mean'),
                           )).style.background_gradient(cmap='RdYlGn', axis=None)
```

---

## 2. Heatmap for Columns

```python
(reduced_retail.pivot_table(index='family', 
                           columns='store_nbr', 
                           values='sales',
                           aggfunc=('mean'),
                           )).style.background_gradient(cmap='RdYlGn', axis=1)
```

---

## 3. Heatmap for Rows

```python
(reduced_retail.pivot_table(index='family', 
                           columns='store_nbr', 
                           values='sales',
                           aggfunc=('mean'),
                           )).style.background_gradient(cmap='RdYlGn', axis=0)
```
