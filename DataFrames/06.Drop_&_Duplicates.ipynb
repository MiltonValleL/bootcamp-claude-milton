# IDENTIFYING & DROPPING DUPLICATES

---

```python
import pandas as pd
import numpy as np
```

```python
product_df = pd.DataFrame(data={'product':['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                                'price':[2.56, 2.56, 4.55, 2.74, 5.44]})
product_df
```

```python
product_df.nunique()
```

```python
# Checking Duplicates
product_df.duplicated()
```

```python
# Checking Duplicates on 'product' column
product_df.duplicated(subset='product')
```

```python
# Dropping the duplicated row
product_df.drop_duplicates()
```

```python
# Dropping duplicated data only from column 'product', keeping last duplicated row and reseting index
product_df.drop_duplicates(subset='product', # select only this column
                           keep='last', # keep the last duplicated row
                           ignore_index=True) # resets the index
```

```python
oil = pd.read_csv('Data/oil.csv')
oil.columns = ['date', 'price']
oil['euro_price'] = oil['price'] * 1.1

# now we'll create a duplicated row from the last row using concat
oil = pd.concat([oil, oil.iloc[[-1]]], ignore_index=True) # usa doble[[]]
oil.tail()
```

```python
# now we'll duplicate the last row using '.loc[]'
oil.loc[len(oil)] = oil.iloc[-1].values
oil.tail()
```

```python
oil.shape
```

```python
oil.nunique(dropna=False)
```

```python
oil.duplicated().sum()
```

```python
oil.drop_duplicates(keep='last', ignore_index=True, inplace=True)
```

```python
import pandas as pd
import numpy as np

oil = pd.DataFrame({
    "fecha":  ["2024-01-01", "2024-01-02", "2024-01-03"],
    "precio": [75.0, 76.5, 77.2],
    "volumen": [1000, 1100, 1050]
})
```

```python
# Caso A: CON .values
oil_a = oil.copy()
oil_a.loc[len(oil_a)] = oil_a.iloc[-1].values
# Resultado: ['2024-01-03', 77.2, 1050] asignado POR POSICIÓN
oil_a
```

```python
# Caso B: SIN .values
oil_b = oil.copy()
oil_b.loc[len(oil_b)] = oil_b.iloc[-1]
# Resultado: Series con index ['fecha', 'precio', 'volumen'] 
# asignada POR NOMBRE DE COLUMNA
oil_b
```
