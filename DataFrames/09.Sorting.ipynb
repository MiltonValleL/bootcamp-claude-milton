# SORTING DATAFRAMES BY INDICES

---

```python
import pandas as pd
import numpy as np
```

```python
retail_df = pd.read_csv('Data/retail_2016_2017.csv')
retail_df.head()
```

```python
condition = retail_df.family.isin(['BEVERAGES', 'DELI', 'DAIRY'])

sample_df = retail_df[condition].sample(5, random_state=2021)
sample_df
```

---

## Sorting Indexm

```python
# sorting by index downwards
sample_df.sort_index(ascending=False)
```

```python
# sorting by column index upwards
sample_df.sort_index(axis=1)
```

---

# Sorting Values

```python
# Sorting one Column
sample_df.sort_values(by='store_nbr')
```

```python
# Sorting more than one column
sample_df.sort_values(by=['family', 'sales'], ascending=[True, False])
```
