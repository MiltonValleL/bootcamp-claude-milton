# THE `.agg()` METHOD & NAMED AGGREGATIONS

---

>The `.agg()` method lets you perform multiple aggregations on a `groupby` object

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head()
```

---

## 1. Single Calculation with `.agg()` Method

```python
single_calc = retail.groupby(by=['store_nbr', 'family'])[['sales', 'onpromotion']].agg('sum').round()
single_calc
```

---

## 2. Multiple Calculation with `.agg()` Method

```python
multiple_calc = retail.groupby(['store_nbr', 'family'])[['sales', 'onpromotion']].agg(['sum', 'mean']).round()
multiple_calc
```

---

## 3. Specific Aggregations by Column

```python
spec_agg = (retail
    .groupby(['store_nbr', 'family'])
    .agg({'sales':['sum','mean', 'std'],
          'onpromotion':['min', 'mean', 'max']})
    .round())

spec_agg
```

---

## 4. Name Aggregated Columns


```python
name_agg = (retail
    .groupby(['store_nbr', 'family'])
    .agg(sales_sum=('sales', 'sum'),
         sales_avg=('sales', 'mean'),
         sales_std=('sales', 'std'),
         on_promotion_max=('onpromotion', 'max'),
         on_promotion_mean=('onpromotion', 'mean'))
).round()
name_agg
```

## LABORATORY

```python
premier_league = pd.read_excel('Data/premier_league_games_full.xlsx')
premier_league.head(3)
```

```python
premier_grouped = (premier_league
    .groupby(['season', 'HomeTeam'], as_index=False)
    .agg({'HomeGoals':['sum', 'mean'],
          'AwayGoals':['sum', 'mean']})
)
premier_grouped
```

```python
name_grouped = (premier_league
    .groupby(['season', 'HomeTeam'], as_index=False)
    .agg(
        homegoal_mean=('HomeGoals', 'mean'),
        homegoal_sum=('HomeGoals', 'sum'),
        AwayGoal_mean=('AwayGoals', 'mean'),
        AwayGoal_sum=('AwayGoals', 'sum')
    ).round()
)
name_grouped
```

## ASSIGNMENT: THE AGG METHOD


```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date'])

transactions = transactions.assign(
    target_pct = transactions.transactions / 2500,
    met_target = transactions.transactions / 2500 >= 1,
    bonus_payable = (transactions.transactions / 2500 >= 1) * 100,
    month = transactions.date.dt.month,
    day_of_week = transactions.date.dt.day_of_week
)

transactions.head()
```

### 1. By Store

```python
(transactions
    .groupby('store_nbr')
    .agg({'met_target':'mean', 'bonus_payable':'sum'})
    .sort_values(by='bonus_payable', ascending=False)
).head()
```

### 2. By Month

```python
(transactions
 .groupby('month')
 .agg({'met_target':'mean', 'bonus_payable':'sum'})
 .sort_values(by='bonus_payable', ascending=False)    
).head()
```

### 3. By Weekday

```python
(transactions
    .groupby('day_of_week')
    .agg({'met_target':'mean', 'bonus_payable':'sum'})
    .sort_values(by='bonus_payable', ascending=False)
).head()
```
