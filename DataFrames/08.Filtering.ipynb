---

````python
df
````

````python
df.query('family in ["CLEANING", "DAIRY"] and sales > 0')
````

---

We can reference variables using `@` symbol

````python
avg_sales = df.loc[:, 'sales'].mean()
avg_sales
````

````python
df.query("family in ['CLEANING', 'DAIRY'] and sales > @avg_sales").sort_values(by='sales')
````

````python
transactions = pd.read_csv('Data/transactions.csv')
transactions.head()
````

````python
# Calculate the percentage of times all stores had more than 2000 transactions
(transactions.transactions > 2000).mean()
````

````python
# Calculate the percentage of times store 25 had more than 2000 transactions, AND 
# calculate the sum of transactions on these days
num = (transactions.query("store_nbr == 25 and transactions > 2000")['transactions'].count())
den = (transactions.query("store_nbr == 25")['transactions'].count())

result = num / den
result
````

````python
# total number of transactions more than 2000 for store 25
transactions.query("store_nbr == 25 and transactions > 2000")['transactions'].sum()
````

````python
# Sum the transactions for stores 25 and 31, that occurred in May or June, and had less than 2000 transactions
transactions.date = pd.to_datetime(transactions.date)
transactions.info()
````

````python
transactions.query("store_nbr in [25, 31] and (date.dt.month == 5 or date.dt.month == 6) and transactions < 2000")['transactions'].sum()
````
