```python
oil = pd.read_csv('Data/oil.csv')
oil.columns = ['date', 'price']
oil['benchmark'] = 90

oil['date'] = pd.to_datetime(oil['date'])
oil['month'] = oil['date'].dt.month
oil.sort_values(by=['date', 'price'], ascending=[True, False], inplace=True)
oil
```

```python
oil['benchmark_ratio'] = oil.loc[:, 'price']/ oil.loc[:, 'benchmark']
oil
```

```python
oil['buy'] = oil.loc[:, 'benchmark_ratio'] < 0.8
oil
```

```python
transactions = pd.read_csv('Data/transactions.csv')
transactions.rename(columns={'store_nbr':'store_number',
                             'transactions':'transaction_count'}, 
                    inplace=True)
transactions.head(3)
```

```python
transactions['pct_target'] = transactions['transaction_count']/2500
transactions['met_target'] = transactions['pct_target'].apply(lambda x: True if x >= 1 else False)
transactions['bonus_payable'] = transactions['met_target'].apply(lambda x: 100 if x == True else 0)

transactions['date'] = pd.to_datetime(transactions['date'])
transactions['month'] = transactions['date'].dt.month
transactions['day_of_week'] = transactions['date'].dt.dayofweek
transactions
```

```python
print(f'Bonus Payable = {transactions['bonus_payable'].sum()}')
```
