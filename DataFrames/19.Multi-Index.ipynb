# MULTI-INDEX DATAFRAMES

<br>

---

>`Multi-Index DataFrames` are generally created through aggregation operations. 
>   - They are stored as a list of tuples, with an item for each layer of the index.

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head()
```

```python
sales_sum = retail.groupby(['family', 'store_nbr'])[['sales']].sum().round(2)
sales_sum
```

```python
sales_sum.index
```

---

## 1. Accessing Rows via `outer index`

```python
sales_sum.loc['AUTOMOTIVE'].head()
```

```python
sales_sum.loc['AUTOMOTIVE':'BEAUTY']
```

---

## 2. Accessing Rows via `Outer` & `Inner` Indices

```python
# One value of the tuple
sales_sum.loc[('AUTOMOTIVE', 5), :]
```

```python
# A range of values from the tuple
sales_sum.loc[('AUTOMOTIVE', 48):('BEAUTY', 2), :]
```

## LABORATORY

```python
premier_league = pd.read_excel('Data/premier_league_games_full.xlsx')
premier_league.head()
```

```python
premier_league.info()
```

```python
agg_premier_league = premier_league.groupby(['season', 'HomeTeam'])[['HomeGoals']].sum()
agg_premier_league.head()
```

```python
# Selecting a single value - Method 1
agg_premier_league.loc['2008/2009', 'Arsenal']
```

```python
# Selecting a single value - Method 2
agg_premier_league.loc[('2008/2009', 'Arsenal')]
```

```python
# Selecting a range of values from a Multi-Index DF (Method 1)
agg_premier_league.loc['2008/2009':'2010/2011']
```

```python
# Selecting a ranges of values from a Multi-Index DF (Method 2)
agg_premier_league.loc[('2008/2009', 'Arsenal'):('2008/2009', 'Chelsea')]
```

```python
# Selecting a range of values from a Multi-Index DF (Methos 3 - .iloc[])
agg_premier_league.iloc[3:7]
```

---

## Aggregations with Multiple Calculations

```python
agg_league = premier_league.groupby(['season', 'HomeTeam']).agg({'HomeGoals':['sum', 'mean', 'min', 'max']})
agg_league
```

```python
# access values with '.iloc[]'
agg_league.iloc[1,3]
```

```python
agg_league.loc[('2008/2009', 'Arsenal'), ('HomeGoals', 'max')]
```

---

```python
agg_league.iloc[3,1]
```

```python
agg_league.loc[('2008/2009', 'Bolton Wanderers'), ('HomeGoals', 'mean')]
```
