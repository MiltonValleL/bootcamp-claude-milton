# ADVANCED CONDITONAL COLUMNS WITH SELECT

---

>NumPy's `select()` function lets you create columns based on multiple conditions.
>* This is more flexible than `np.where()` function or `pd.where()` method

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Data/retail_2016_2017.csv')
df['tax_amount'] = df['sales'] * 0.05
df['total'] = df['sales'] + df['tax_amount']
df['taxable_category'] = df['family'] != 'BABY CARE'
df
```

```python
conditions = [
    (df['date'] == '2017-02-23') & (df['family'] == 'BABY CARE'),
    (df['date'] == '2016-12-24') & (df['family'] == 'BOOKS'),
    (df['date'] == '2016-09-06') & (df['store_nbr'] > 28),
]

choices = ['Winter Clearance', 'Christmas Eve', 'New Store Special']

df['Sale_Name'] = np.select(condlist=conditions, choicelist=choices, default='No Sale')
df
```

```python
df['Sale_Name'].value_counts()
```

```python
oil = pd.read_csv('Data/oil.csv')
oil.columns = ['date', 'price']
oil.head()
```

```python
oil1 = oil.copy()
oil1['buy'] = np.where(oil1['price'] > 100, 'Too High', 'Buy')
oil1
```

```python
oil1['buy'].value_counts()
```

---

```python
conditions =[
    (oil.price > 100),
    (oil.price <= 100) & (oil.price > 50),
    (oil.price < 50)
]

choices = ["Don't Buy", "Buy", "Strong Buy"]

oil['buy'] = np.select(condlist=conditions, choicelist=choices, default='Missing')
oil.head()
```

```python
oil.tail()
```

```python
oil.buy.value_counts()
```

```python
transactions = pd.read_csv('Data/transactions.csv')
transactions.head()
```

```python
transactions['date'] = pd.to_datetime(transactions['date'])
transactions.info()
```

---

Create a ‘seasonal_bonus’ column that applies to these dates:

    * All days in December: Call the December bonus `Holiday Bonus`
    * Sundays in May: Call the May bonus `Corporate Month`
    * Mondays in July: Call the August bonus ‘Summer Special’

If no bonus applies, the column should display `None`. Finally, calculate the total bonus owed at $100 per day.

---

```python
# The Complete Week Index: Monday=0, Tuesday=1, Wednesday=2, Thursday=3, Friday=4, Saturday=5, Sunday=6
# thos numbers work for "dt.dayofweek" or "dt.day_of_week"

conditions = [
    (transactions['date'].dt.month == 12),
    (transactions['date'].dt.month == 5) & (transactions['date'].dt.dayofweek == 6),
    (transactions['date'].dt.month == 7) & (transactions['date'].dt.day_of_week == 0)
]

choices = ['Holiday Bonus', 'Corporate Month', 'Summer Special']

transactions['seasonal_bonus'] = np.select(condlist=conditions, choicelist=choices, default='None')
transactions.head()
```

```python
transactions['seasonal_bonus'].value_counts()
```

```python
(transactions['seasonal_bonus'] != 'None').sum()*100
```
