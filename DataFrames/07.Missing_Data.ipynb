# MISSING DATA

---

```python
import pandas as pd
import numpy as np
```

```python
product_df = pd.DataFrame({'product':[pd.NA, 'Dairy', 'Dairy', pd.NA, 'Fruits'],
                           'price':[2.56, pd.NA, 4.55, 2.74, pd.NA],
                           'product_id':[1, 2, 3, 4, 5]})
product_df
```

```python
# Counting total Missing Values
product_df.isna().sum()
```

```python
product_df.isnull().sum()
```

```python
# Other way to see if your DataFrame has Missing Values, is by using the Method ".info()"
product_df.info()
```

---

Like with `Series`, the `.dropna()` and `.fillna()` methods let you **handle missing data** in a DataFrame by either removing them or replacing them with other values

```python
product_df
```

```python
# FutureWarning: Downcasting object dtype arrays on .fillna, .ffill, .bfill is deprecated and will change in a future version. 
# Call result.infer_objects(copy=False) instead. To opt-in to the future behavior, 
# set `pd.set_option('future.no_silent_downcasting', True)`
pd.set_option('future.no_silent_downcasting', True)

# Replacing all missing values in one line
product_df.fillna(0)
```

```python
# Replacing Nan Values per column
product_df.fillna({'price': 0})
```

# ASSIGNMENT: MISSING DATA

---

```python
oil = pd.read_csv('Data/oil.csv')
oil.head()
```

---

```python
# Checking if there are missing values
oil.isna().sum()
```

```python
oil_zero = oil.fillna(0)
print(f'The Mean of Oil Price filled with 0 is: {oil_zero['dcoilwtico'].mean()}')
```

```python
mean_val = oil['dcoilwtico'].mean()
oil_mean = oil.fillna(mean_val)
print(f'The Mean of Oil Price filled with the Mean value is: {oil_mean['dcoilwtico'].mean()}')
```
