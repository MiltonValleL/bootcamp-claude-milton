# Key Takeaways

>1. Always answer the `3 key questions` to choose the right visual
>>* What type of data are you working with?
>>* What do you want to communicate?
>>* Who is the end user?

>2. Do `NOT` prioritize variety over effectiveness
>>* Choose Chart types based on how clearly they communicate the data underneath - You can customize later.

>3.  Eliminate noise and distractions to **facilitate understanding**
>>* Perfection is achieved not when there is nothing more to add, but when there is nothing left to take away

>4.  **Tell a story** with the data to guide the user to the insights
>>* Use titles, strategic labels, and callouts to create a clear narrative

## Plotting Methods

**1. PyPlot API**

Charts are created with the `plot()` function, and modified with additional functions

```python
plt.plot(y);
```

**2. Object-Oriented Method**

Charts are created by defining a `plot object`, and modified using `figure` & `axis` methods. 

- This method gives a lot more control over what's going on inside figures and allows for extreme customization.

```python
# Create a figure object and assign it to the 'fig' variable
fig = plt.figure()

# Add a chart, or axis object, to the figure and assign it to the 'ax' variable
ax = fig.add_subplot()

# Call the axis plot() method to draw the chart
ax.plot(y);
```

```python
fig, ax = plt.subplots()
ax.plot(z)
fig.suptitle("OVERALL FIGURE TITLE")
ax.set_title("Chart Title");
```

```python
plt.plot(z)
plt.title("This is the Title")
plt.show()
```

```python
# Step by Step plotting
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(z)
ax.set_title("This the Title")
fig.suptitle("OVERALL TITLE")
```

```python
# Resumed plotting
fig, ax = plt.subplots()
ax.plot(z)
ax.set_title("This is the second title")
fig.suptitle("SECOND OVERALL TITLE")
plt.show()
```

## Plotting DataFrames


```python
data_dict = {
    "Sales": [10, 20, 30, 40],
    "Profits": [2, 4, 3, 1],
    "Date": ["2022-0101", "2022-02-01", "2022-03-01", "2022-04-01"],
}

df = pd.DataFrame(data=data_dict, columns=["Sales", "Profits"], index=data_dict["Date"])
df
```

```python
# Using a DataFrame the index is the X-axis, and the columns are the Y-axis
fig, ax = plt.subplots()
ax.plot(df)
ax.set_title("Product Sales and Profits")
plt.show()
```

```python
# To improve customization it's better to create a single plot for each DataFrame column
fig, ax = plt.subplots()
ax.plot(df.index, df["Sales"], ls="-.")
ax.plot(df.index, df["Profits"], ls="--")
ax.set_title("Product Sales and Profits II")
plt.show()
```
