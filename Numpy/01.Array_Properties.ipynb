# Numpy Arrays & Array Properties

```python
import numpy as np
```

```python
a = np.array(range(5))
a
```

```python
# Create 2D Array
b = np.array([range(5), range(5)])
b
```

```python
# Add an array with an int
b + 2
```

```python
print(f"Shape of array a: {a.shape}   -   Shape of array b: {b.shape}")
```

```python
print(f"Size of array a: {a.size}   -   Size of array b: {b.size}")
```

```python
print(
    f"Number of Dimensions of array a: {a.ndim}   -   Number of Dimensions of array b: {b.ndim}"
)
```

```python
print(f"Data Type of array a: {a.dtype}   -   Data Type of array b: {b.dtype}")
```

```python
c = np.array(["I", "Love", "Python"])
print(f"Shape: {c.shape}\nSize: {c.size}\nndim: {c.ndim}\ndtype: {c.dtype}")
```

## Assignment: Array Basics

```python
my_list = [x * 10 for x in range(1, 11)]
my_list
```

```python
my_array = np.array(my_list)
my_array
```

```python
print(
    f"ndim: {my_array.ndim}\nshape: {my_array.shape}\nsize: {my_array.size}\ndtype: {my_array.dtype}"
)
```

```python
my_array2 = my_array.reshape(5, 2)
my_array2
```

```python
print(
    f"ndim: {my_array2.ndim}\nshape: {my_array2.shape}\nsize: {my_array2.size}\ndtype: {my_array2.dtype}"
)
```
