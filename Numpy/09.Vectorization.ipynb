# VECTORIZATION & BROADCASTING

>`Vectorization` is the process of pushing array operations into optimized C code, which is easier and more efficient than writting **for loops**.


---





```python
import numpy as np
import time
```

```python
def for_loop_multiply_lists(list1, list2):
    product_list = []
    for elem1, elem2 in zip(list1, list2):
        product_list.append(elem1 * elem2)
    return product_list


def multiply_arrays(array1, array2):
    return array1 * array2
```

---

```python
lst1 = list(range(1000))
lst2 = list(range(1000))
```

```python
%%timeit -r 5 -n 10000
for_loop_multiply_lists(lst1, lst2)
```

```python
%%timeit
for_loop_multiply_lists(lst1, lst2)
```

---

```python
np_1 = np.array(lst1)
np_2 = np.array(lst2)
```

```python
%%timeit -r 5 -n 10000
multiply_arrays(np_1, np_2)
```

```python
%%timeit
multiply_arrays(np_1, np_2)
```

>`PRO TIP:`
>Use vectorized operations whenever possible when manipulating data, and avoid writing loops

# BROADCASTING

>`Broadcasting` lets you perform vectorized operations with arrays of different sizes, where NumPy will expand the smaller array to "fit" the larger one.

---



```python
test_array = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
test_array
```

#### Example 1

```python
# Braadcasting with a single number [1 x 1]
test_array + 1
```

#### Example 2

```python
# Broadcasting with a row [1 x 3]
test_array + np.array([3, 2, 1])
```

#### Example 3

```python
# Broadcasting with a columns [3 x 1]
test_array + np.array([3, 2, 1]).reshape(3, 1)
```

#### Example 4 (Error)

```python
# Be careful with the size of each array. Becuase Broadcast can expand every randomly to fullfill blank spaces of an Array
test_array + np.array([2, 1])
```

#### Example 5

```python
a = test_array[0, :]
b = test_array[:, 1]

print(f"Array a: {a}   -   Array b: {b}")
```

```python
c = b.reshape(3, 1)
c
```

```python
# Example No. 4 - Sum rows with columns
a + c
```

#### NOTE: Broadcasting can be implemented with 1D, 2D, 3D, nD

```python
dim2 = np.ones((2, 3, 3), dtype=int)
dim2
```

```python
dim1 = np.ones(3, dtype=int)
dim1
```

# ASSIGNMENT: Vectorization & Broadcasting

```python
import pandas as pd
```

```python
retail_df = pd.read_csv(
    "Data/retail_2016_2017.csv", skiprows=range(1, 11000), nrows=1000
)
retail_df.sample(10)
```

```python
family_array = np.array(retail_df["family"])
sales_array = np.array(retail_df["sales"])

print(f"Length of Family Array: {len(family_array)}")
```

```python
produce_array = sales_array[family_array == "PRODUCE"]
print(f'Total length of "produce_array": {len(produce_array)}\n')
produce_array
```

```python
rng = np.random.default_rng(2022)

random_array = rng.random(30)
random_array
```

```python
produce_sample = produce_array[random_array < 0.5]
print(f'Total length of "produce_sample": {len(produce_sample)}\n')
produce_sample
```

```python
mean = produce_sample.mean()
mean
```

```python
median = np.median(produce_sample)
median
```

```python
np.where(
    produce_sample < median,
    "below_both",
    np.where(produce_sample > mean, "above_both", "above_median"),
)
```
