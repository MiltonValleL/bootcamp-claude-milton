```python
import numpy as np
```

```python
sales = [[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]]

sales_array = np.array(sales)
sales_array
```

```python
# Add array + 2
sales_array + 2
```

## Laboratory: Array Operations

```python
rng = np.random.default_rng(616)

inventory = rng.integers(0, 100, 10)
inventory
```

```python
inventory - 24
```

```python
inventory * inventory[::-1]
```

```python
inventory / 2
```

```python
price = (rng.random(10) * 10).round(2)
price
```

```python
price * inventory
```

```python
(price * inventory).sum()
```

```python
inventory_list = list(inventory)
inventory_list
```

```python
new_inventory = []

for x in inventory_list:
    new_inventory.append(int(x) + 2)

new_inventory
```

```python
[int(x) + 2 for x in inventory]
```

```python
[round(float(x * y), 1) for x, y in zip(inventory_list, price)]
```

## ASSIGNMENT 4: Array Operation

```python
prices = np.array([5.99, 6.99, 22.49, 99.99, 4.99, 49.99])

total = prices + 5
total
```

```python
discount_pct = np.random.default_rng(2022).random(6).reshape(1, 6)
discount_pct
```

```python
pct_owed = 1 - discount_pct
pct_owed
```

```python
final_owed = np.round(pct_owed * total, 2)
final_owed
```
