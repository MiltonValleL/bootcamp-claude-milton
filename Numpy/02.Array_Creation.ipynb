### Random Number Generation

#### np.default_rng(seed)

```python
# Creates a random number generator
from numpy.random import default_rng

rng = default_rng(12345)
```

#### np.random(n)

```python
# Returns 'n' random numbers from a uniform distribution between 0 and 1
rand_array = rng.random(10)
rand_array
```

#### rng.normal(mean, stdev, n)

```python
# Returns "n" random numbers from a normal distribution with a given mean and standard deviation+
rng = default_rng(42)
mean, stdev = (5, 1)

random_normal = rng.normal(mean, stdev, size=10)
random_normal
```

```python
rng = default_rng(12345)
mean, stdev = (5, 1)

random_normal = rng.normal(mean, stdev, size=10)
random_normal
```

---


>PRO TIP: 
>
>Eventhough it's optional, make sure to **set a seed** when generating random numbers to ensure you and others can recreate the work you have done (*the value for the seed is less important*)

```python
# Generate 100 random and integer numbers fro 0 to 10
rng.integers(0, 10, 100)
```

```python
# Generate 10 random and normally ditribute numbers
# with a Mean = 50, and a Std = 5
rng.normal(50, 5, 10)
```
