# INDEXING & SLICING ARRAYS

```python
import numpy as np
```

```python
product_array = np.array(
    [
        "fruits",
        "vegetables",
        "cereal",
        "dairy",
        "eggs",
        "snacks",
        "beverages",
        "coffe",
        "tea",
        "spices",
    ]
)
product_array
```

```python
# get 2nd item
product_array[1]
```

```python
# get last item
product_array[-1]
```

```python
# Slices
product_array[:5]
```

```python
# more Slices with step
product_array[5::2]
```

---

### Slicing 2-Dimensional Array

```python
product_array_2D = product_array.reshape(2, 5)
product_array_2D
```

```python
# Select "Coffe" - Method 1
product_array_2D[1][2]
```

```python
# Select "Coffe" - Method 2
product_array_2D[1, 2]
```

```python
# Now select all Rows but only the last 3 columns
product_array_2D[:, 2:]
```

```python
# Now select only the 2nd Row with all elements
product_array_2D[1, :]
```

## Laboratory: Indexing & Slicing

```python
int_array = np.arange(12)
int_array
```

```python
int_array[2]
```

```python
int_array[3:-3]
```

```python
int_array[:]
```

```python
int_array[::2]
```

### Reshaping Original 1D Array into a 2D Array

```python
new_int = int_array.reshape(3, 4)
new_int
```

```python
new_int[1, :]
```

```python
new_int[:2, 2:]
```

```python
new_int[:, :]
```

```python
new_int[2, 1]
```
