## LABORATORY: Array Aggregations

```python
rng = np.random.default_rng(616)

price = (rng.random(10) * 10).round(2)
price
```

```python
inventory = rng.integers(0, 100, 10)
inventory
```

```python
inventory.mean()
```

```python
inventory.std()
```

```python
price * inventory
```

```python
total_sales = price * inventory
total_sales_sum = total_sales.sum()
total_sales_sum
```

```python
total_sales_mean = total_sales.mean()
total_sales_mean
```

```python
total_sales_max = total_sales.max()
total_sales_max
```

```python
# Index of the Max Value - "total_sales"
total_sales.argmax()
```

```python
# Index of the Min Value - "total_sales"
total_sales.argmin()
```

```python
price_2d = price.reshape(5, 2)
price_2d
```

```python
# Sum across rows
price_2d.sum(axis=0)
```

```python
# Sum across columns
price_2d.sum(axis=1)
```
