# WHERE FUNCTION

> The `where()` NumPy function peforms a logical test and returns a given value if the test is **True**, or another if the test is **False**.

```python
import numpy as np
```

```python
inventory_array = np.array([12, 102, 18, 0, 0])

product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs"])
```

```python
np.where(inventory_array <= 0, "Out of Stock", "In Stock")
```

```python
# Methos 2: Complex Way
np.where(inventory_array <= 0, "Out of Stock", product_array)
```

## LABORATORY: Where Function

```python
my_array = np.arange(20)
my_array
```

```python
np.where(my_array % 2 == 0, "EVEN", "odd")
```

```python
# Using double np.where()
np.where(my_array % 2 == 0, "EVEN", np.where(my_array == 9, my_array, "odd"))
```

```python
# Using double logical evaluation AND double np.where()
np.where(
    my_array % 2 == 0,
    "EVEN",
    np.where((my_array == 9) | (my_array == 3), my_array, "odd"),
)
```

```python
# Provided array of ages
ages = np.array([25, 12, 15, 64, 35, 80, 45, 10, 22, 55])
```

```python
# Use NumPy's where function to find indices of ages that are 18 or older
adult_ages = np.where(ages >= 18)
adult_ages
```

# ASSIGNMENT 5: Filtering Arrays

>Filter the product array to only include those with prices greater than 25.
>
>Modify your logic to include 'cola', despite it not having a price greater than 25. Store the elements returned in an array called `fancy_feast_special`.
>
>Next, create a shipping cost array where the cost is 0 if price is greater than 20, and 5 if not.

```python
products = np.array(
    ["salad", "bread", "mustard", "rare tomato", "cola", "gourmet ice cream"]
)

prices = np.array([5.99, 6.99, 22.49, 99.99, 4.99, 49.99])
print(products, prices)
```

```python
products[prices > 25]
```

```python
mask = (prices > 25) | (products == "cola")

fancy_feast_special = products[mask]
fancy_feast_special
```

```python
shipping_cost = np.where(prices > 20, 0, 5)
shipping_cost
```
