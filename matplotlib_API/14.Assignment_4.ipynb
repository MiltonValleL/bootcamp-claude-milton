```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date'])

df_assign = (transactions
    .query("store_nbr in [2, 3]")
    .pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum')
)

df_assign
```

```python
sns.set_style('whitegrid')

df_assign.plot.scatter(x=3, 
                       y=2,
                       title='Scatterplot between Stores 2 and 3',
                       xlabel='Store No. 3',
                       ylabel='Store No. 2')
```

```python
sns.set_style('darkgrid')

df_assign.plot.scatter(x=3, 
                       y=2,
                       c=df_assign.index.month,
                       colormap='Set1',
                       title='Scatterplot between Stores 2 and 3',
                       xlabel='Store No. 3',
                       ylabel='Store No. 2')
```
