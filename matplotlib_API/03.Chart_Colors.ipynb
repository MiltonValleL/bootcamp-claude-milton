# CHART COLORS

---

You can modify the **chart formatting** by using `.plot()` method arguments

>---
>- `title` = Title to use for the chart
>- `xlabel` = Name to use for the x-axis
>- `ylabel` = Name to use for the y-axis
>- `color` = **“color”** or **“#hexacode”** – Color(s) to use for the data series
>- `cmap` = **“color palette”** – preset color palette to apply to the chart
>- `style` = **“symbol”** – Style for the line (dashed, dotted, etc.)
>- `legend` = **True/False** – Adds or removes the legend from the chart
>- `rot` = **degrees** – Degrees to rotate the x-axis labels for readability
>- `figsize` = **(width, height)** – Size for the chart in inches
>- `grid` = **True/False** – Adds gridlines to the chart when True
>- `subplots` = **True/False** – Creates a separate chart for each series when True
>   - `sharex` = **True/False** – Each series in the subplot shares the x-axis when True
>   - `sharey` = **True/False** – Each series in the subplot shares the y-axis when True
>   - `layout` = **(rows, columns)** – the rows and columns to break the subplots int

```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv', parse_dates=['date'])
retail.head(3)
```

```python
(retail
    .query("family == 'AUTOMOTIVE' and date.dt.year == 2017 and date.dt.month == 3 and store_nbr == 12")
    .set_index('date')
    .plot(y='sales',
         title = 'Automotive - March Total Sales on Store No. 12',
         xlabel = 'Date',
         ylabel = 'Total Sales in USD',
         color = 'gray')
)
```

```python
# Multiple Family Products to plot
(retail.query("family in ['AUTOMOTIVE', 'BEAUTY', 'PERSONAL CARE'] and store_nbr == 30 and date.dt.month == 12")
    .pivot_table(index='date', columns='family', values='sales')
    .plot(
        title='Total Sales for Automotive, Beauty & Personal Care on Store No. 30',
        xlabel='Date', ylabel='Total Sales [USD]',
        color=['black', 'red', 'green']
         )
)
```

```python
# Multiple Family Products to plot
(retail.query("family in ['AUTOMOTIVE', 'BEAUTY', 'PERSONAL CARE'] and store_nbr == 30 and date.dt.month == 12")
    .pivot_table(index='date', columns='family', values='sales')
    .plot(
        title='Total Sales for Automotive, Beauty & Personal Care on Store No. 30',
        xlabel='Date', ylabel='Total Sales [USD]',
        color=['#000000', '#FF0000', '#0000ff']
         )
);
```

## Using `alpha` argument

>When there are overlapping line plots is a good practice to adjust the `alpha` argument.
>
>Let's see the following plot without using the `alpha` argument.

```python
stores = transactions.query("store_nbr in [15, 16, 17]").pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum')
stores.plot(title='Total Sales for Stores 15, 30 and 45',
            xlabel='DATE',
            ylabel='Total Number of Transactions',
            color=['blue', 'orange', 'r'],
            );
```

---

>Now using the `alpha` argument...

```python
stores.plot(title='Total Sales for Stores 15, 30 and 45',
            xlabel='DATE',
            ylabel='Total Number of Transactions',
            color=['blue', 'orange', 'r'],
            alpha = 0.36
            );
```
