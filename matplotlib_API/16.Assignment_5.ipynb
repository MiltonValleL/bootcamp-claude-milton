```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date'])

df_assignment = (transactions
    .pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum'))
df_assignment
```

```python
sns.set_style('darkgrid')

df_assignment.loc[:, [2, 3]].plot.hist(
    bins=10,
    alpha=0.5);
```
