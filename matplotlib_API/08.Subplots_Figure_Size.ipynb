# SUBPLOTS & FIGURE SIZE

---

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
retail=pd.read_csv('Data/retail_2016_2017.csv', parse_dates=['date'])

retail.query("family in ['AUTOMOTIVE', 'BEAUTY', 'HARDWARE', 'BOOKS'] and date.dt.month == 3 and date.dt.year == 2017", inplace=True)
retail
```

```python
df_plot = retail.pivot_table(index='date', columns='family', values='sales', aggfunc='mean')

df_plot.plot(title='Total Sales on Abril 2017',
             xlabel='DATE',
             subplots=True);
```

```python
df_plot.plot(
    subplots=True,
    layout=(2,2),
    sharey=True,
    legend=False,
    grid=True,
    title=['Automotive', 'Beauty', 'BOOKS', 'Hardware']
);
```

```python
sns.set_style('darkgrid')

df_plot.plot(
    figsize=(15, 6),
    subplots=True,
    layout=(2,2),
    sharey=False,
    legend=True,
    title=['Automotive', 'Beauty', 'Books', 'Hardware'],
);
```

## LABORATORY

```python
transactions = pd.read_csv('Data/transactions.csv',parse_dates=['date']).set_index('date')
transactions
```

```python
df_plot = (transactions
    .query("store_nbr in [44, 47] and date.dt.year == 2014")
    .pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='mean')
)

sns.set_style('white')

df_plot.plot(
    figsize=(15, 6),
    subplots=True,
    layout=(1,2),
    sharex=True,
    sharey=True,
    title='Store 44 and 47 Transaccions 2014',
    xlabel='Date format',
    ylabel='Daily Transactions', 
    color=['#187878', '#c507f0'],
    grid=True,
    alpha=0.45,
    style=['.-', '.-'],
);
```
