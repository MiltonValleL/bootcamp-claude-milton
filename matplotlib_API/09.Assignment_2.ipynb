```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date'])

transactions = (transactions
    .query("store_nbr in [1, 2, 3, 4]")
    .pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum')
)

transactions
```

```python
sns.set_style('darkgrid')

transactions.plot(
    figsize=(18, 9),
    subplots=True,
    layout=(2,2),
    sharex=True,
    sharey=True,
    title='Total Transactions for Stores 1, 2, 3 and 4',
    xlabel='DATE',
    ylabel='Transaction [USD]',
    grid=True,
    # color=['blue','orange', 'green', 'brown'],
);
```
